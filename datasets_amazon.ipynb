```python
import pandas as pd
mx_devices = pd.read_csv('/home/miguel-oscar/DS_IBM_Course/DS_IBM_Course/Tarea_2-Maestria/best_sellers_amazon/mexico/csv/mx-amazon-devices.csv', sep='|', index_col=0, on_bad_lines='skip')

mx_devices.head()
```

```python
mx_devices.shape
```

```python
mx_devices.columns
```

```python
mx_devices.info()
```

```python
# Se eliman las columnas que no tienen datos
mx_devices = mx_devices.dropna(axis=1, how='all')

mx_devices.info()
```

```python
# Revisamos los valores nulos de la columna 'Product names'
mx_devices[mx_devices['Product Names'].isnull()]
```

```python
#### SE PUEDEN ELIMINAR LAS FILAS CON VALORES NULOS EN LA COLUMNA ANTERIOR PORQUE NO APORTAN INFORMACION RELEVALENTE PARA EL ANALISIS
mx_devices.dropna(subset=['Product Names'], inplace=True)
```

```python
mx_devices.info()
```

```python
# Revisamos los valores nulos de la columna 'Stars'
mx_devices[mx_devices['Stars'].isnull()]
```

```python
#### SE PUEDEN ELIMINAR LAS FILAS CON VALORES NULOS EN LA COLUMNA ANTERIOR PORQUE NO APORTAN INFORMACION RELEVALENTE PARA EL ANALISIS
mx_devices.dropna(subset=['Stars'], inplace=True)
mx_devices.info()
```

```python
# Cambiamos los tipos de las columnas 'Rank' y 'Reviews' a INT y el de 'time' a tipo fecha
# No debe de haber errores porque estas columnas no contienen valores nulos
mx_devices['time'] = pd.to_datetime(mx_devices['time'], format='%Y-%m-%d %H:%M')
mx_devices['Rank'] = mx_devices['Rank'].astype(int)
mx_devices['Reviews'] = mx_devices['Reviews'].astype(int)
```

```python
mx_devices.info()
```

```python
# Estadísticas descriptivas de las variables numéricas
mx_devices.describe().T
```

```python
# Se separan las temporalidades de las variables de tiempo y se crean nuevas variables para el análisis.
mx_devices['hours'] = mx_devices['time'].dt.hour
mx_devices['date'] = mx_devices['time'].dt.date
mx_devices['weekday'] = mx_devices['time'].dt.day_name()
mx_devices['time_hour'] = mx_devices['time'].dt.floor('h')

mx_devices.head()
```

### Análisis de la cantidad de registros por día

```python
conteos_diarios = mx_devices.groupby('date').size()

conteos_diarios.plot(kind='line', marker='o', title='Registros por dia')
plt.xticks(rotation=45)
plt.show()
```

### Análisis de los de reviews promedio por día

```python
reviews_diarios = mx_devices.groupby('date')['Reviews'].mean()

reviews_diarios.plot(kind='line', marker='o', title='Reviews promedio por dia')
plt.xticks(rotation=45)
plt.show()
```

### Análisis del promedio de Strellas por hora del día

```python
estrellas_diarias = mx_devices.groupby('hours')['Stars'].mean().reset_index()

sns.lineplot(data=estrellas_diarias, x='hours', y='Stars', markers='o')
plt.title('Promedio de estrellas por hora del dia')
plt.xticks(rotation=45)
plt.show()
```

### Análisis del promedio de Strellas día de la semana

```python
# Define el orden correcto de los dias de la semana
orden_dias = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

estrellas_diarias = (mx_devices.groupby('weekday')['Stars']
                     .mean()
                     .reindex(orden_dias)  # Reordenar según nuestra lista
                     .reset_index())

sns.lineplot(data=estrellas_diarias, x='weekday', y='Stars', markers='o')
plt.title('Promedio de estrellas por hora del dia')
plt.xticks(rotation=45)
plt.show()
```

## Análisis de productos

```python
# Estadísticas descriptivas de las variables categóricas
mx_devices.describe(include='object')
```

```python
# Top 10 de productos por reviews
top_products = (mx_devices.groupby('Product Names')['Reviews']
                .sum()
                .sort_values(ascending=True)
                .head(10))
plt.figure(figsize=(12, 10))
top_products.plot(kind='barh', title='Top 10 de productos por reviews')
plt.tight_layout()
plt.show()
```

```python
# Agrupar por día y hora y ordenar descendentemente por la cantidad
# mx_devices.groupby(mx_devices['time'].dt.floor('h')).size().sort_values(ascending=False)
grouped_by_date = (mx_devices.groupby(mx_devices['time'].dt.floor('h'), as_index=False)
 .size()
 #.sort_values(by= 'size', ascending=False)
 .rename(columns={'size': 'cantidad'})
)
grouped_by_date
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
```

```python
sns.lineplot(data=grouped_by_date, x='time', y='cantidad', marker='o')
plt.title('Cantidad de registros por hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=mx_devices, x='time', bins=30)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
