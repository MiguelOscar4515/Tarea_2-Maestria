## K-Means Clustering en el dataset de best sellers de Amazon

```python
# Definir las caracteristicas a usar para el clustering
features = ['Rank', 'Stars', 'Reviews', 'Price_std_or_min']
```

```python
# Eliminar los nulos
df_kmeans = mx_devices[['Rank', 'Stars', 'Reviews', 'Price_std_or_min']].dropna()
```

```python
# Hacer escalamiento
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(df_kmeans)
```

```python
# Metodo del codo

from sklearn.cluster import KMeans
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_scaled)
    inertia.append(kmeans.inertia_)
    
plt.plot(K, inertia, marker='o')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Metodo del codo')
plt.show()
```

### Tomamos k=4 basado en el codo anterior porque es el valor en el que la disminución de la inercia comienza a ser menos pronunciada.

```python
# Se ejecuta K-Means con k=4
kmeans = KMeans(n_clusters=4, random_state=42)
df_kmeans['cluster'] = kmeans.fit_predict(x_scaled)
```

```python
df_kmeans.groupby('cluster').mean()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(x_scaled)

df_kmeans['pc1'] = components[:,0]
df_kmeans['pc2'] = components[:,1]

plt.scatter(df_kmeans['pc1'], df_kmeans['pc2'], c=df_kmeans['cluster'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clusters de productos (PCA)')
plt.show()
```

## Interpretación de los clusters
### Se ejecuta K-Means con k=4

1. Cluster morado oscuro (abajo a la izquierda, PC1 muy negativo y PC2 muy negativo) <br>
   Este grupo está súper separado del resto de los cluster, lo que indica:

   * Productos muy diferentes al resto

   * Estos perfiles producen clusters como:

      - Precios muy altos o muy bajos

      -Rank muy malo (alta numeración)

      -Reviews muy bajas

      - Stars extremas (muy buenas o muy malas)

* Este PCA muestra este cluster muy aislado, por lo que es un buen cadidato para ser un cluster de outliers o categoría distinta.

2. Cluster rojo (centro pero un poco hacia abajo)
   * Este cluster está cerca del centro, ligeramente más disperso y con un PC2 ligeramente negativo. Usualmente representa:

   * Productos promedio en todo:

      - Rank medio

      - Reviews medias

      - Precio promedio

      - Stars alrededor del promedio

      - Mercado competitivo

* Son productos que representa "lo promedio" o "lo típico"

3. Cluster naranja (rosa) claro (a la derecha, PC1 positivo)
   * Este clúster se separa hacia el lado positivo de PC1.

   * Eso típicamente significa:

      - Productos con características más altas

      - Reviews más altas

      - Precio más alto

      - Rank más bajo (mejor)

      - Stars ligeramente superiores

   * Productos más exitosos / premium.

4. Cluster negro (arriba, PC2 muy positivo)
   * Este grupo se verticaliza hacia arriba, lo que significa que:

      - Productos con muchas reviews pero Rank inconsistent

      o

      - Productos que tienen altos Stars pero bajo volumen

   * PCA suele separar columnas como "Reviews" y "Stars" en direcciones diferentes, es por eso que este cluster está arriba, porque tiene alguna variable muy positiva, pero no necesariamente todas.

   * Puede ser:

      - Productos con buenas calificaciones

      - Reviews crecientes en el tiempo

      - Precios bajos o muy estables
